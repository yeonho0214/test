```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('winequality-red.csv')
# 중복된 행 제거
df_no_duplicates = df.drop_duplicates()
```

```python
# Isolation Forest 모델 학습
isolation_forest_model = IsolationForest(contamination=0.1)
isolation_forest_model.fit(df_no_duplicates)
```

```python
# 이상치 여부 확인
outliers = isolation_forest_model.predict(df_no_duplicates) == -1
```

```python
# 결과 시각화
plt.scatter(df_no_duplicates.iloc[:, 0], df_no_duplicates.iloc[:, 1], c=outliers, cmap='viridis', alpha=0.5)
plt.title('Isolation Forest Outlier Detection (After Removing Duplicates)')
plt.show()
```

```python
plt.scatter(df_no_duplicates.iloc[:, 0], df_no_duplicates.iloc[:, 1], c='orange', alpha=0.5)
plt.title('Isolation Forest Outlier Detection (After Removing Duplicates)')
plt.show()
```

```python
# 결과 시각화 - 색상 코드 사용
plt.scatter(df_no_duplicates.iloc[:, 0], df_no_duplicates.iloc[:, 1], c='#FF5733', alpha=0.5)
plt.title('Isolation Forest Outlier Detection (After Removing Duplicates)')
plt.show()
```

```python
# 결과 시각화 - 이상치 색상 변경
plt.scatter(df_no_duplicates.iloc[:, 0], df_no_duplicates.iloc[:, 1], c=np.where(outliers, 'red', 'blue'), alpha=0.5)
plt.title('Isolation Forest Outlier Detection (After Removing Duplicates)')
plt.show()
```
